```python
df2=df1.drop(['Family'],axis=1) #very high vif value and high skewness with outliers
```

# Removing outliers using zscore

```python
#Removing outliers using zscore
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df2))
```

```python
df_new=df2[(z<3).all(axis=1)]
```

```python
df2.shape
```

```python
df_new.shape
```

```python
#divide dataframe into features and label
x=df_new.drop(['Happiness Score'],axis=1)
y=df_new['Happiness Score']
```

# Removing skewness from data

```python
x.skew()
```

```python
#removing complete skewness
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
```

```python
lr=LinearRegression()
```

```python
#splitting the dataframe
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(pred_train,y_train)*100,1)==round(r2_score(pred_test,y_test)*100,1):
        print('At random state',i,'the model performs verywell')
        print('At random state',i)
        print("the training accuacy is",r2_score(pred_train,y_train)*100)
        print("the testing accuacy is",r2_score(pred_test,y_test)*100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=8)
lr.fit(x_train,y_train)
pred_test=lr.predict(x_test)
pred_train=lr.predict(x_train)
```

```python
mean_absolute_error(y_test,pred_test)
```

```python
np.sqrt(mean_squared_error(y_test,pred_test))
```

```python
r2_score(y_test,pred_test)
```

```python
r2_score(y_train,pred_train)
```
